```
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```
# scatter diagram
np.random.seed(0)

# x-axis data
x = np.random.randn(30)

# y-axis data
y = np.sin(x) + np.random.randn(30)

# graph size
plt.figure(figsize=(15,6))

# graph drawing
#plt.plot(x,y,'o')

# scatter diagram
plt.scatter(x,y)

# graph title
plt.title('Scatter diagram')
# x-axis name
plt.xlabel('X')
# y-axis name
plt.ylabel('Y')
# graph grid
plt.grid(True)
```

```
# continuous curve

np.random.seed(0)

# data range
numpy_data_x = np.arange(1000)

# random number
numpy_random_data_y = np.random.randn(1000).cumsum()

# graph size
plt.figure(figsize=(15,6))

# graph drawing
plt.plot(numpy_data_x, numpy_random_data_y, label = 'data1')
plt.legend()

plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
```

```
# graph size
plt.figure(figsize=(15,6))

# 2-lines 1-row graph : No.1 graph
plt.subplot(2,1,1)
x = np.linspace(-10,10,100)
plt.plot(x, np.sin(x))

# 2-lines 1-row graph : No.2 graph
plt.subplot(2,1,2)
y = np.linspace(-10,10,100)
plt.plot(y,np.sin(2*y))

plt.grid(True)
```

```
def my_function(x):
    return x**2 + 2*x + 1

x = np.arange(-10,10)
plt.figure(figsize=(15,6))
plt.plot(x, my_function(x))
plt.grid(True)
```

```
# uniform random number
x = np.random.uniform(0.0, 1.0, 10000)
y = np.random.uniform(0.0, 1.0, 10000)

plt.figure(figsize=(5,5))
plt.plot(x,y,'o')

cnt = 0
x1 = []
y1 = []
for i in range(0,10000):
    if np.math.hypot(x[i],y[i]) < 1:
        cnt += 1
        x1.append(x[i])
        y1.append(y[i])

print(cnt)

plt.plot(x1,y1,'o')
plt.grid(True)

print(4*cnt/(10000))
```
